```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df = df.set_index('Date')
```

```python
print(plt.style.available)
```

```python
df.plot(kind = 'line' , title = 'Ice creame Rating' , xlabel = 'Daily rating', ylabel = 'Scores')
```

```python
df.plot.barh( stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating' , y = 'Overall Rating' , s = 5 , c ='red')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.pie(y = 'Flavor Rating')
```
